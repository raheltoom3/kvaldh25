```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Tekst
text = """
Preili Iseseisev Stella Terna Ma haaran kaks keefiri pakki ja käekoti sülle, kui vaatad mind korraks ja siis ütled mulle: "Puhka jalga, preili iseseisev kõike kandma ei pea üksi ma tean et oled harjund ei sinu jalas püksid, kuid lase ma aitan, kui mu pead õrnalt paitad." Ja sa vist ise ka ei tea, et koorem mida lahustad ei koosne piimatoodete raskusest või võtmetest mu taskutes. Vaid aastate jagu ise hakkama saamist, kui vastutuse merel üksi aerutanud paadis. See on vist vanimate tütarde saatuslik sündroom, kus mehed on oht ja isa on fantoom, kus printsessile andmata jäi talle kuuluv kroon ja siin su ees ma seisan nagu ohutules loom. "Puhka jalga, preili boss, ma tulen ja ma teen võtan tulelt potud, mis hinges üle on keend." Ja ma seisan tardunult, jookseb elu silme ees, et ma ei ole teadnud, kes on turvaline mees, mitte lihtsalt minu väljamõeldud fantaasia hüpotees vaid keegi kes maandab elu ümber ja mu sees. Mõistan, et pean vabaks andma kaitsed, mida loonud need, mis mind on elusana tänasesse toonud. Kuid rohkem ma ei soovi, ei saa ega taha, iseseisvuse rüü ma auga matan maha. Sind, mu kallis preili, kes rinda aina tagus sõnu "ma saan üksi" su sõnastikus jagus teistelegi jagada. Uskudes et survivalmodes jääme kindlalt elama, kuid nukker oleks jääda seda paati üksi vedama. Nendel matustel ma naeratan, annan kergusega maale kui enda sees sean paika polaarsuse kaale. Ma tahan olla naine, kes vastu sinu kaljut peksab kirglikult kui laine. Elujoovastusest purjus, sest sina oled kaine ma saan olla müstika sest sina oled maine. Ma toetan, siis su õlale oma koormatud pea ja hingan välja mõttega, et "nii on päris hea"
"""

# Kõik väiketähtedeks
text_lower = text.lower()

# Stopwords set
stopwords = set(["ja", "kui", "oled", "mida", "mu", "kus", "kuid", "ees",
                 "pakki", "su", "ma", "on", "et", "ei", "siis", "kes", "mis", "või", "sest", "see", "ka"])
stopwords_lower = set(word.lower() for word in stopwords)

# WordCloudi loomine
wc = WordCloud(
    width=1920,
    height=1080,
    background_color="white",
    max_font_size=200,
    colormap="Reds",
    stopwords=stopwords_lower
).generate(text_lower)

# Kuvamine ja salvestamine
plt.figure(figsize=(19.2, 10.8))  # Full HD proportsioonid
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.savefig("wordcloud_hd.png", dpi=300)  # kõrge kvaliteet
plt.show()
```
